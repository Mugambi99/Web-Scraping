```python
import random

# Constants for the board size and starting position
BOARD_SIZE = 100
START_POSITION = 0

# Constants for the snakes and ladders positions
SNAKES = {
    99: 70,
    93: 64,
    87: 24,
    62: 18,
    54: 34,
    19: 7
}
LADDERS = {
    98: 78,
    95: 75,
    92: 88,
    47: 26,
    16: 6
}

def move(position, dice_roll):
    """
    Move the player's position on the board based on the dice roll.
    """
    position += dice_roll
    if position in SNAKES:
        print(f"Oops! You landed on a snake and slid down to {SNAKES[position]}.")
        position = SNAKES[position]
    elif position in LADDERS:
        print(f"Yay! You climbed up a ladder to {LADDERS[position]}.")
        position = LADDERS[position]
    return position

def play_game():
    """
    Play the Snakes and Ladders game.
    """
    position = START_POSITION
    while position < BOARD_SIZE:
        print(f"You are currently at {position}.")
        dice_roll = random.randint(1, 6)
        print(f"You rolled a {dice_roll}.")
        position = move(position, dice_roll)
    print("You won the game!")

if __name__ == "__main__":
    play_game()
```
